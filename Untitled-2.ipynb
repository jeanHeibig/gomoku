```python
from backend.players.bots.alpha_beta import find_best_move
from backend.board import prettyprint
```

```python
def bb(s):
    return np.uint64(eval("0b"+s[::-1].replace('\n', '').replace(' ', '')))
```

```python
def toggle(x, y, t):
    if t:
        return y, x
    return x, y
```

```python
t = 1
```

```python
m = find_best_move(
    *toggle(bb("""
        00000000
        00000010
        00000100
        00101000
        00010000
        00010100
        00000000
        00000000
    """),
    bb("""
        00000001
        00000000
        00010000
        00010100
        00001000
        00101000
        00000000
        00000000
    """), t),
    1, 3
)
t = 1 - t
prettyprint(m)
```
